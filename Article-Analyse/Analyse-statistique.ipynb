```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import os
%matplotlib inline 
print("Dossier actuel :", os.getcwd())
print("Fichiers/Dossiers ici :", os.listdir())
try:

    df_life_expectancy = pd.read_csv('../Dataset/Economie/life expectancy.csv')    
    print(f"Données chargées avec succès : {len(df_life_expectancy)} lignes et {len(df_life_expectancy.columns)} colonnes.")
    df_Global_Economy_Indicators = pd.read_csv('../Dataset/Economie/Global Economy Indicators.csv')
    print(f"Données chargées avec succès : {len(df_Global_Economy_Indicators)} lignes et {len(df_Global_Economy_Indicators.columns)} colonnes.")

    display(df_life_expectancy.head(5)) 
    display(df_Global_Economy_Indicators.head(5)) 

except FileNotFoundError:
    print("Erreur : Le fichier est introuvable. Vérifiez le chemin d'accès : './Dataset/Economie/life expectancy.csv'")
except Exception as e:
    print(f"Une erreur s'est produite : {e}")
```

```python
corr_matrix = df_life_expectancy.corr(numeric_only=True)
    
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matrice de Corrélation - Espérance de Vie et Facteurs')
plt.show()


```

```python
corr_matrix = df_Global_Economy_Indicators.corr(numeric_only=True)
    
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matrice de Corrélation - Global Economy Indicators')
plt.show()
```
